## Cómo trabajar con subplots

Subplot<strong>s</strong> es como Subplot pero se diferencian en que el primero se vale de las bondades de trabajar con el método orientado a objetos de Matplotlib

```python
# Importar
import matplotlib.pyplot as plt
import numpy as np

# Crear variables
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
# Crear una gráfica tipo Fig
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

```python
# Crear varias gráficas tipo Fig
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y,'b')
axes[1].plot(y, x,'r')
plt.show()
```

```python
# Crear una matriz de tipo Fig, con dos filas y cuatro columnas
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x), 'y')
axes[0, 2].plot(x, np.tan(x), 'r')
axes[0, 3].plot(x, np.cos(x)**2)
fig.tight_layout() # Mejora la visualización de los ejes de cada gráfico
```

```python
# Forma alternativa para generar una matriz de tipo Fig, accediendo a cada gráfica mediante los nombres dados en una tupla
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x)**2)
fig.tight_layout()
# El resultado es el mismo que en el ejemplo anterior
```
