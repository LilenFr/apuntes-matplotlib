## Contexto de gráficos

Para dar <strong>contexto</strong> a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas

```python
# Importar
import matplotlib.pyplot as plt
import numpy as np

# Crear variables
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
# Crear una gráfica común
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.show()
```

```python
# Contexto: Títulos
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title('Relacion X - Y') # <--

axes[1].plot(y, x)
axes[1].set_title('Relacion Y - X') # <--
plt.show()
```

```python
# Contexto: Etiquetas
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X') # <--
axes[0].set_ylabel('Y') # <--

axes[1].plot(y, x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y') # <--
axes[1].set_ylabel('X') # <--
plt.show()
```

```python
# Contexto: Tamaño
fig, axes = plt.subplots(1, 2, figsize=(5, 5)) # <--
axes[0].plot(x, y)
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y, x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
plt.show()
```

```python
# Contexto: Leyendas
fig, axes = plt.subplots(1, 2, figsize=(5, 5))
#axes[0].plot(x, y, label='sin(x)') # <--
axes[0].plot(x, y, label='$sin(x)$') # <-- Para dejar notación matemática
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend() # <-- Para que aparezca en el gráfico

axes[1].plot(y, x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
plt.show()
```

```python
# Hacer el gráfico anterior pero con plotly
plt.plot(x, y, label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
