## Subplot

A través de una matriz de gráficos, Subplot permite crear gráficos dentro de una gráfica y se puede acceder a ellos por sus índices.

```python
# Importar
import matplotlib.pyplot as plt
import numpy as np

# Crear variables
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# Usar subplot para crear una matriz de gráficos de una fila y dos columnas

# Gráfico 1
plt.subplot(1, 2, 1) # Parámetros: número de filas, número de columnas, primer index
plt.plot(x, y, 'r--') 
# Gráfico 2 
plt.subplot(1, 2, 2) # Parámetros: númweo de filas, número de columnas, segundo index
plt.pie(y)
plt.show()
```

```python
# Agregar diferentes Plot al gráfico
# Gráfico 1
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--') # Plot 1
plt.plot(y, x, 'b:') # Plot 2
#Gráfico 2
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```
