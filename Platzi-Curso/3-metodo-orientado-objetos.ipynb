## Método orientado a objetos

El método <strong>orientado a objetos</strong> es otra manera de hacer gráficas. Se diferencia de Plotly en que tiene mayor personalización, involucra más código y es más amigable a múltiples diagramas.

Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gráficas(axes), de las cuales cada una posee sus propios ejes(axis).

![Alt text](https://static.platzi.com/media/user_upload/Captura-845012cb-41c1-4c4b-bad5-427afbf00703.jpg)

```python
# Importaciones
import matplotlib.pyplot as plt
import numpy as np

# Crear variables
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# Gráficar con el método orientado a objetos
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.5, 0.9]) # Parámetros: posición eje x, posición eje y, tamaño eje x, tamaño eje y
axes.plot(x, y, 'b')
plt.show()
```

```python
# Podemos además crear varias gráficas dentro del lienzo
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = figura.add_axes([0.17, 0.55, 0.4, 0.3])
axes.plot(x, y, 'b')
axes2.plot(y, x, 'r:')
axes2.set_facecolor('grey')
plt.show()
```
